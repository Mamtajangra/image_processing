```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
img = imread("image\lion.jpg")
print(img)
```

```python
plt.imshow(img)
plt.title("original image")
plt.show()
```

grayscale 

```python
# Formula: Gray = 0.2989*R + 0.5870*G + 0.1140*B
gray = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(gray,cmap = "gray")
plt.title("grayscale")
plt.show()
```

inverted

```python
invert = 255 - img[:,:,:3]
plt.imshow(invert)
plt.title("invert")
plt.show()
```
